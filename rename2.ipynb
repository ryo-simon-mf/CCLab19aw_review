```python
import glob
import os
import shutil
```

```python
od_dir = '/Users/ryosimon/Documents/CCLab/fashsionSorting/pics'
```

```python
base_dir = '/Users/ryosimon/Documents/CCLab/fashsionSorting/pics_rename2'
os.mkdir(base_dir)
```

```python
# renamed techno punkrock hiphop directory

techno_dir = os.path.join(base_dir, 'techno')
os.mkdir(techno_dir)
punkrock_dir = os.path.join(base_dir, 'punkrock')
os.mkdir(punkrock_dir)
hiphop_dir = os.path.join(base_dir, 'hiphop')
os.mkdir(hiphop_dir)
```

```python
# original techno, punkrock, hiphop files
fb_techno = os.path.join(od_dir, 'techno')
fb_punkrock = os.path.join(od_dir, 'punkrock')
fb_hiphop = os.path.join(od_dir, 'hiphop')
```

```python
te_files = os.listdir(fb_techno)  
te_count = len(te_files)

pu_files = os.listdir(fb_punkrock)  
pu_count = len(pu_files) -1

hi_files = os.listdir(fb_hiphop)  
hi_count = len(hi_files) -1


print(te_count)  
print(pu_count)  
print(hi_count)  
```

```python
t_path = os.path.join(fb_techno, '*')
tlist = glob.glob(t_path)
print("before")
print(tlist)
ti = 0

for tfile in tlist:
    now = techno_dir + '/techno.' + str(ti) + '.jpg'
    os.rename(tfile, now)
    ti = ti + 1

talist = glob.glob(techno_dir)
print('after')
print(talist)
```

```python
p_path = os.path.join(fb_punkrock, '*')
plist = glob.glob(p_path)
print("before")
print(plist)
pi = 0

for pfile in plist:
    now = punkrock_dir + '/punkrock.' + str(pi) + '.jpg'
    os.rename(pfile, now)
    pi = pi + 1

palist = glob.glob(punkrock_dir)
print('after')
print(palist)
```

```python
h_path = os.path.join(fb_hiphop, '*')
hlist = glob.glob(h_path)
print("before")
print(hlist)
hi = 0

for hfile in hlist:
    now = hiphop_dir + '/hiphop.' + str(hi) + '.jpg'
    os.rename(hfile, now)
    hi = hi + 1

halist = glob.glob(hiphop_dir)
print('after')
print(halist)
```
